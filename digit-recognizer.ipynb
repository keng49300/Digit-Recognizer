```python
import pandas as pd
data=pd.read_csv("/kaggle/input/digit-recognizer/train.csv")
X=data.drop(columns='label')
Y=data['label']
test_data=pd.read_csv("/kaggle/input/digit-recognizer/test.csv")
test_data
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=1)                    ## define the nearest centroid model
model.fit(X, Y)                               ## train the model with the dataset
```

```python
label  = model.predict(test_data)            ## predict a label for the example
import matplotlib.pyplot as plt, matplotlib.image as mpimg
from sklearn.model_selection import train_test_split
from sklearn import svm
%matplotlib inline
# i=500
# img=test_data.iloc[i].values
# img=img.reshape((28,28))
# plt.imshow(img,cmap='gray')

for i in range(len(test_data)):
    plt.title(label[i])
    img = test_data.iloc[i].values
    img = img.reshape((28, 28))
    plt.imshow(img, cmap='gray')
    plt.show()
    
```
